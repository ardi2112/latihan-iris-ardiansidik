IMPORT LIBRARY YANG DIBUTUHKAN

```python
from __future__ import print_function

import IPython
print('IPython:', IPython.__version__)

import numpy
print('Numpy:', numpy.__version__)

import pandas
print('Pandas:', pandas.__version__)

import sklearn
print('Scikit-learn:', sklearn.__version__)
```

**LOAD DATA IRIS** bawaan asli (built-in) dari modul sklearn.datasets di dalam library scikit-learn

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

MENAMPILKAN TIPE DATA NUMPY ARRAY DAN PANDAS DATAFRAME

```python
X, y = iris.data, iris.target
print(X[:5])
print(y[:5])
```

MEMBUAT PIPELINE UNTUK NUMPY DAN PANDAS

```python
import pandas as pd

df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target #target

df.head()

from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC

pipe_numpy = make_pipeline(StandardScaler(), LinearSVC())
pipe_pandas = make_pipeline(StandardScaler(), LinearSVC())
```

MELAKUKAN SPLITTING

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,  random_state=46)

df_train, df_test = train_test_split(df, test_size=0.2, random_state=46)
```

MELAKUKAN TRAINING

```python
#Numpy dilakukan scaling dan model -> kunci untuk training adalah nama pipeline.fit
pipe_numpy.fit(X_train, y_train)
y_pred = pipe_numpy.predict(X_test)

#Menampilkan hasil evaluasi hasil dari model yang sudah dibuat
print(classification_report(y_test, y_pred))

#Pandas
X_train_df, y_train_df = df_train.drop('target', axis=1), df_train['target']
X_test_df, y_test_df = df_test.drop('target', axis=1), df_test['target']
pipe_pandas.fit(X_train_df, y_train_df)
y_pred_df = pipe_pandas.predict(X_test_df)

print("Hasil evaluasi model pandas")
print(classification_report(y_test_df, y_pred_df))
```

SAVE MODEL

```python
import pickle

with open('model_numpy.pkl', 'wb') as f:
  pickle.dump(pipe_numpy, f)

with open('model_pandas.pkl', 'wb') as model_file:
  pickle.dump(pipe_pandas, model_file)
```

LOAD MODEL

```python
#Numpy
with open('model_numpy.pkl', 'rb') as model_file:
  loaded_model_numpy = pickle.load(model_file)

#Pandas
with open('model_pandas.pkl', 'rb') as model_file:
  loaded_model_pandas = pickle.load(model_file)
```

MEMASUKKAN DATA PREDIKSI BARU UNTUK MENCOBA

```python
new_data = [[2,3,4,5]]
loaded_model_numpy.predict(new_data)
```

```python
new_data_df = pd.DataFrame([[1,1,1,1]], columns=iris.feature_names)
loaded_model_pandas.predict(new_data_df)
```
